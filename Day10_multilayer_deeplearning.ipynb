```python
import numpy as np
from sklearn.neural_network import MLPClassifier

# dataset
X = np.array([
    [2,15],
    [3,20],
    [8,60],
    [9,75]
])

y = np.array([0,0,1,1])

# multilayer neural network
model = MLPClassifier(hidden_layer_sizes=(3,),
                      activation='relu',
                      max_iter=500)

model.fit(X,y)

# prediction
pred = model.predict([[7,55]])

print("Prediction:",pred)
```
